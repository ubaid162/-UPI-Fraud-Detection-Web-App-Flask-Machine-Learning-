Feature Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
encoders = {}

# List of categorical columns to encode
cols_to_encode = ['MerchantCategory', 'TransactionType', 'BankName']

# Loop through each column
for col in cols_to_encode:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    encoders[col] = le  # Save encoder for future use
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['1/day','3/day','5/day']])
```

```python
Transaction_frequency = oe.fit_transform(data[['TransactionFrequency']])
data['Transaction_frequency'] = Transaction_frequency.astype('int')
```

```python
data.drop(['TransactionFrequency'],axis = 1,inplace= True)
```

Feature selection

```python
x = data.drop(['FraudFlag'],axis = 1)
y = data[['FraudFlag']]
```

Splitting data into test and train

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.2,random_state=42)
```

Train the Model

1 . Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,classification_report
```

```python
rfc = RandomForestClassifier(n_estimators=3)
rfc.fit(x_train,y_train)
```

```python
y_pred = rfc.predict(x_test)
y_pred
```

```python
rfc_acc = accuracy_score(y_test,y_pred)
print(rfc_acc)
```

```python
rf_conf = confusion_matrix(y_test,y_pred)
rf_conf
```

2. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_pred = lr.predict(x_test)
```

```python
lr_acc = accuracy_score(y_test,lr_pred)
print(lr_acc)
```

```python
lr_conf = confusion_matrix(y_test,lr_pred)
lr_conf
```
